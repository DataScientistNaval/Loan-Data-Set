```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('applicant.csv')
data.head(5)
```

```python
data2=pd.read_csv('loan.csv')
data2.head(5)
```

## Merging the datasets

```python
mergeddf=pd.merge(data,data2,on='applicant_id',how='outer')
mergeddf.head(5)
```

```python
mergeddf.drop(['applicant_id','Telephone','loan_application_id'],axis=1,inplace=True)
```

# Conclusion = Every feature is almost evenly distributed for High and Low risk, so we cannot conclude that any single feature should be used for creditworthiness of the applicant.

### Let try to study low risk applicant and make the best analysis out of them.

```python
G1=mergeddf[mergeddf.high_risk_applicant==0] #give all records with low risk applicant
G1
```

```python
G1.groupby('Loan_history').size() 
#people with critical/pending loans at other banks and existing loans paid back duly till now are at Low risk.
#And we have considered them because we have less data for others.
```

```python
G1.groupby('Housing').size()
#people with own house are more on the safer side for creditworthiness
```

```python
G1.groupby('Number_of_dependents').size()
#People with 1 dependent are more likely to be creditworthy according to the data.
```

```python
G1.groupby('Purpose').size()
#We can consider FF&E,electronic equipment,new vehicle.
```

```python
G1.groupby('Savings_account_balance').size()
#does not matter if the person has low or high saving account balance. As we can clearly see people with low balance are also at low risk.
#though Saving account has lot of null values but if we fill it with the mode we will get again low balance as the most, so we are not filling it for now.
```

```python
G1.groupby('Property').size()
#almost equal distribution - cannot conclude anything.
```

```python
G1.groupby('Foreign_worker').size()
#so here we can see most of the applicant who are at low risk are Foreign workers. #yay we will consider them.
```

```python
G1.groupby('Marital_status').size()
# we can clearly see that single person are at low risk. We will consider them as well.
```

```python
G1.groupby('Has_been_employed_for_at_least').size()
#cannot conclude.
```

```python
G1.groupby('Has_been_employed_for_at_most').size()
#cannot conclude.
```

```python
G1.groupby('Has_coapplicant').size()
# more applicant with 0 coapplicant are on the low risk. We will consider them.
```

```python
G1.groupby('Has_guarantor').size()
#more applicant with 0 guarantor are on the low risk. We will consider them.
```

```python
G1.groupby('Number_of_existing_loans_at_this_bank').size()
#if a person has 1 or 2 existing loan are on the safe side that is Low risk. We will consider them as well.
```

```python
#We will not consider Other EMI plans because it has a lot of null values in it.
```
