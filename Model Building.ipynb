```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve

#importing the required libraries.
```

```python
scaler=MinMaxScaler()
scaleddata=scaler.fit_transform(finaldata.drop('high_risk_applicant',axis=1))
#transforming the independent features
```

```python
scaleddata.shape
```

```python
X=scaleddata
y=finaldata['high_risk_applicant']
```

```python
def classify(model,X,y):
    Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=15)
    model.fit(Xtrain,ytrain)
    ypredicted=model.predict(Xtest)
    print(metrics.accuracy_score(ytest,ypredicted))
    print(metrics.confusion_matrix(ytest,ypredicted))
    print(classification_report(ytest, ypredicted))
    crossval=cross_val_score(model,X,y,cv=5)
    print(np.mean(crossval))
        

#creating a function to classify and making cross validation also
#cross validation score can be seen at bottom of every model
```

```python
def fittingcheck(model,X,y):
    Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=15)
    model.fit(Xtrain,ytrain)
    ytrainpredict=model.predict(Xtrain)
    print(metrics.accuracy_score(ytrain,ytrainpredict))
    
# function to check overfitting and underfitting of the model
# making sure that the model is generalised
# Training Model accuracy
```

## Logistic Reg

```python
classify(LogisticRegression(),X,y)
```

```python
fittingcheck(LogisticRegression(),X,y)
# The model is generalised.
```

## SVC Model

```python
classify(SVC(),X,y)
```

```python
fittingcheck(SVC(),X,y)
# It is a overfitting model i.e, Low bias and High Variance.
```

## Naive Bayes

```python
classify(GaussianNB(),X,y)
```

```python
fittingcheck(GaussianNB(),X,y)
```

### Among all the models the best one is Naive Bayes because it has less false negative than SVC and Logistic Regression.
### Naive Bayes is also a generalised model.
### SVC is slightly overfitting.
### Logistic Regression is a generalised model.
